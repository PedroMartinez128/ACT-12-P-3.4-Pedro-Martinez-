## Las variables
1. CT---> Costo del traje.
2. DE---> Descuento que se obtendra.
3. PF---> Precio final del traje.

---
## Las formulas
1. $DE = CT * 0.15$
2. $DE = CT * 0.08$
3. $PF = CT - DE$

```python
# Para cuando es Verdadero(SI)
# Calculando el descuento usando if else
# Datos de entrada: CT (Costo del Traje) DE (Descuento), PF (Precio Final)
# Mostrar el descuento es
# El descuento final es
CT = float(input("Ingrese el costo del traje: "))
if CT > 2500:
    DE = CT * 0.15
else:
    DE = CT * 0.08
PF = CT - DE
print("El descuento es: ", DE)
print("El precio final es: ", PF)
print("Programa realizado por Pedro Martinez NL = 34")
```

```python
# Para cuando es Falso(NO)
# Calculando el descuento usando if else
# Datos de entrada: CT (Costo del Traje)
# Datos de salida: DE (Descuento), PF (Precio Final)
CT = float(input("Ingrese el costo del traje: "))
if CT > 2500:
    DE = CT * 0.15
else:
    DE = CT * 0.08
PF = CT - DE
print("El descuento es: ", DE)
print("El precio final es: ", PF)

print("Programa realizado por Pedro Martinez NL = 34")
```
